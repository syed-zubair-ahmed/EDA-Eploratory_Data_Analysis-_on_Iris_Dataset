###  Step 5: Bivariate Analysis
  Analyzing the relationship between two variables to detect patterns, correlations, or group separations.

 **Scatter Plot**  
  Used to visualize the relationship between two numerical variables.  

 **Boxplot by Category**  
  Shows how a numerical variable (e.g., sepal_length) varies across categories (e.g., species)


```python
# Scatter plot: Petal Length vs Petal Width, colored by Species
sbn.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris)
plt.title("Scatter Plot - Petal Length vs Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
# Boxplot: Sepal Length grouped by Species
sbn.boxplot(x='species', y='sepal_length', data=iris)
plt.title("Boxplot - Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length")
plt.show()
```

###  Step 6: Multivariate Analysis
  Exploring multiple variables simultaneously to uncover patterns, correlations, or class separability.

- **Pairplot**  
  A grid of scatter plots for every numerical variable combination.  
  Using `hue='species'` helps visualize how different species are distributed across dimensions.

- **Correlation Heatmap**  
  Displays correlation coefficients between numeric variables.  
  `annot=True` shows the exact values; `cmap='coolwarm'` gives a color gradient from negative to positive correlation.


```python
# Pairplot with species hue
sbn.pairplot(iris, hue='species')
plt.suptitle("Pairplot - Relationship Across All Features", y=1.025)
plt.show()
```

```python
# Correlation heatmap
corr_matrix = iris.corr(numeric_only=True)  # In case of categorical columns
sbn.heatmap(corr_matrix, annot=True, cmap='mako', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

### Step 7: Outlier Detection
 Outliers are extreme values that differ significantly from other observations.  
 Boxplots are useful to visually identify these values in numerical features.

- **Boxplot**  
  Here we drop the categorical column `'species'` and plot all numeric columns together.  
  Any points outside the whiskers are considered potential outliers.


```python
# Boxplot for all numeric features to detect outliers
sbn.boxplot(data=iris.drop('species', axis=1), palette='dark')
plt.title("Outlier Detection with Boxplot")
plt.xticks(rotation=45)
plt.show()
```
